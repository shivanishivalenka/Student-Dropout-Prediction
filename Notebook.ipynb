```python
#trimming outliers
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR
    
    # Filter the DataFrame to exclude outliers
    filtered_data = data[(data[column] >= lower_fence) & (data[column] <= upper_fence)]
    
    return filtered_data

outlier_features= ['Age at enrollment',
  'Scholarship holder',
  'Curricular units 1st sem (grade)',
  'Curricular units 2nd sem (grade)',
  'Curricular units 1st sem (approved)',
  'Curricular units 2nd sem (approved)']
# Iterate over the specified columns and remove outliers
for column in outlier_features:
    df = remove_outliers_iqr(df, column)
```

```python
for ever in outlier_features:
    print(ever)
    sns.boxplot(df[ever].values)
    
    plt.title(ever)
    plt.show()
    
```

```python
import pandas as pd

outlier_threshold = 1.5
def cap_outliers(df, feature, threshold):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_fence = Q1 - threshold * IQR
    upper_fence = Q3 + threshold * IQR
    
    # Cap extreme values
    df[feature] = df[feature].clip(lower=lower_fence, upper=upper_fence)
    
    return df
clip_outlier_features = ['Age at enrollment',
  'Curricular units 1st sem (grade)',
  'Curricular units 2nd sem (grade)']
# Iterate over the outlier_features and cap outliers
for feature in clip_outlier_features:
    df = cap_outliers(df, feature, outlier_threshold)
```

# Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
scaler=StandardScaler()
df=df[df.Target!=2]
df["Target"].describe()
df1 = df[['Age at enrollment','Scholarship holder','Gender','Displaced','Tuition fees up to date','GDP','Curricular units 1st sem (grade)'
           ,'Curricular units 2nd sem (grade)','Curricular units 1st sem (approved)','Curricular units 2nd sem (approved)']]
```

```python
X = df1
y = df["Target"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2, random_state = 4)
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
#({'Dropout':0,'Enrolled':1,'Graduate':2})
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```

```python
from sklearn.ensemble import RandomForestClassifier 
model1 = RandomForestClassifier()
model1.fit(X_train,y_train)
```

```python
y_pred1 = model1.predict(X_test) 
```

```python
accuracy = accuracy_score(y_test, y_pred1)
conf_matrix = confusion_matrix(y_test, y_pred1)
classification_rep = classification_report(y_test, y_pred1)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred2 = clf.predict(X_test)
```
